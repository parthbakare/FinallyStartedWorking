The first plot between Feature 1 and Feature 2 classifies the two labels perfectly

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('data.txt', delimiter='\t', skiprows=1)
#count = 1

for a in range(1, 11):
    for b in range(a+1, 11):

        dict1 = {}
        dict2 = {}

        for k in range(0, data.shape[0]):
            if data[k][0] == np.float64(1):
                dict1[data[k][a]] = data[k][b]
        
            else:
                dict2[data[k][a]] = data[k][b]
        
        x1 = []
        y1 = []

        for i in sorted(dict1):
            x1.append(i)
            y1.append(dict1[i])

        x2 = []
        y2 = []

        for i in sorted(dict2):
            x2.append(i)
            y2.append(dict2[i])

        plt.plot(x1, y1, color = 'r', label = "Label 1")
        plt.plot(x2, y2, color = 'b', label = "Label 2")

        plt.title('Feature {} v/s Feature {}'.format(b, a))
        plt.ylabel('Feature {}'.format(b))
        plt.xlabel('Feature {}'.format(a))
        plt.legend()
        plt.show()
        #print(count)
        #count+=1
```
